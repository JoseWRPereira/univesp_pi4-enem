```python
df = pd.DataFrame()
```

```python
dfaux = uDados[['NU_INSCRICAO','SG_UF_PROVA','NO_MUNICIPIO_PROVA']].groupby(['SG_UF_PROVA','NO_MUNICIPIO_PROVA']).count().reset_index()
dfaux
```

```python
df[['uf','cidade','qtd']] = dfaux[['SG_UF_PROVA','NO_MUNICIPIO_PROVA', 'NU_INSCRICAO']]
df
```

```python
dfaux = uDados[['SG_UF_PROVA','NO_MUNICIPIO_PROVA','NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']].groupby(['SG_UF_PROVA','NO_MUNICIPIO_PROVA']).mean().reset_index()
df[['ch_mean', 'lc_mean', 'mt_mean','cn_mean','pt_mean']] = dfaux[['NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']]
df
```

```python
dfaux = uDados[['SG_UF_PROVA','NO_MUNICIPIO_PROVA','NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']].groupby(['SG_UF_PROVA','NO_MUNICIPIO_PROVA']).std().reset_index()
df[['ch_std', 'lc_std', 'mt_std','cn_std','pt_std']] = dfaux[['NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']]
df
```

```python
dfaux = uDados[['SG_UF_PROVA','NO_MUNICIPIO_PROVA','NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']].groupby(['SG_UF_PROVA','NO_MUNICIPIO_PROVA']).min().reset_index()
df[['ch_min', 'lc_min', 'mt_min','cn_min','pt_min']] = dfaux[['NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']]
df
```

```python
dfaux = uDados[['SG_UF_PROVA','NO_MUNICIPIO_PROVA','NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']].groupby(['SG_UF_PROVA','NO_MUNICIPIO_PROVA']).max().reset_index()
df[['ch_max', 'lc_max', 'mt_max','cn_max','pt_max']] = dfaux[['NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']]
df
```

```python
dfaux = uDados[['SG_UF_PROVA','NO_MUNICIPIO_PROVA','NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']].groupby(['SG_UF_PROVA','NO_MUNICIPIO_PROVA']).median().reset_index()
df[['ch_median', 'lc_median', 'mt_median','cn_median','pt_median']] = dfaux[['NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']]
df
```

```python
dfaux = uDados[['SG_UF_PROVA','NO_MUNICIPIO_PROVA','NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']].groupby(['SG_UF_PROVA','NO_MUNICIPIO_PROVA']).quantile(.25).reset_index()
df[['ch_q1', 'lc_q1', 'mt_q1','cn_q1','pt_q1']] = dfaux[['NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']]
df
```

```python
dfaux = uDados[['SG_UF_PROVA','NO_MUNICIPIO_PROVA','NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']].groupby(['SG_UF_PROVA','NO_MUNICIPIO_PROVA']).quantile(.50).reset_index()
df[['ch_q2', 'lc_q2', 'mt_q2','cn_q2','pt_q2']] = dfaux[['NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']]
df
```

```python
dfaux = uDados[['SG_UF_PROVA','NO_MUNICIPIO_PROVA','NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']].groupby(['SG_UF_PROVA','NO_MUNICIPIO_PROVA']).quantile(.75).reset_index()
df[['ch_q3', 'lc_q3', 'mt_q3','cn_q3','pt_q3']] = dfaux[['NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']]
df
```
